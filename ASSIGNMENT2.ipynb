# Learn about magic commands and what they are used  for .

 Magic commands are an enhancement of the interactive IPython kernel (used by Jupyter) and make performing common tasks easier than using Python code alone.

 Magic commands are easily identified because they all start with the % character. Line magics start with a single % and apply to the code on the rest of the line. Cell magics start with %% and apply to the rest of the code in a cell.

 Some common uses of magic commands are

 - test the speed of your code.
 - display the memory usage of your code.
 - Help with autocomplete.
 - Write a code to a file.
 - Reload code from a file.

# What are the difference between a bar chart, count plot and histogram?

A histogram is a data visualization type designed to show the distribution of interval or continuous data. In histograms, data is shown in the form of contiguous bars, where each bar corresponds to a data range or a bin.


A count plot can be thought of as a histogram across a categorical, instead of quantitative, variable. 

Bar plots look similar to count plots, but instead of the count of observations in each category, they show the mean of a quantitative variable among observations in each category. This bar plot uses the tips dataset and shows the average bill paid among people who visited the restaurant on each day of the week.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
sales = [400, 350, 300, 450]

plt.figure(figsize=(6,4))
plt.barh(fruits, sales, color='pink')
plt.title('Fruit Sales', loc='left')
plt.xlabel('Units Sold')
plt.ylabel('Fruits')
plt.tight_layout
plt.show()
```

```python


scores = [55, 62, 67, 70, 72, 75, 78, 80, 82, 85, 88, 90, 92, 95, 98, 100, 100, 65, 77, 89]
plt.figure(figsize=(8,6))
# Create a histogram
plt.hist(scores, bins=8, color='pink', edgecolor='black')

# Add titles and labels
plt.title('Distribution of Math Test Scores')
plt.xlabel('Score Ranges')
plt.ylabel('Number of Students')

# Show the plot
plt.show()
```

```python


# Sample dataset: Heights of students in three classes
class_A = [150, 160, 165, 170, 175, 180, 155, 165, 160, 170]
class_B = [140, 145, 150, 155, 160, 165, 170, 175, 180, 185]
class_C = [155, 160, 165, 170, 175, 180, 185, 190, 195, 200]

# Combine the data into a list
data = [class_A, class_B, class_C]

# Create a boxplot
plt.boxplot(data, tick_labels=['Class A', 'Class B', 'Class C'], patch_artist=True)

# Add titles and labels
plt.title('Height Distribution of Students in Different Classes')
plt.ylabel('Height (cm)')

# Show the plot
plt.show()
```

```python


# Sample dataset: Scores of students in three subjects
data = {
    "Math": [85, 90, 78, 92, 88, 76, 95, 89, 84, 91],
    "Science": [88, 85, 82, 95, 90, 80, 96, 87, 83, 92],
    "English": [78, 82, 85, 88, 84, 79, 91, 86, 80, 87]
}

# Convert the dataset into a NumPy array
scores = np.array([data["Math"], data["Science"], data["English"]])

# Calculate the correlation matrix
correlation_matrix = np.corrcoef(scores)
plt.figure(figsize=(9,8))
# Create a heatmap
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')

# Add color bar
plt.colorbar(label='Correlation Coefficient')

# Add labels
plt.xticks(ticks=[0, 1, 2], labels=["Math", "Science", "English"])
plt.yticks(ticks=[0, 1, 2], labels=["Math", "Science", "English"])
plt.title("Correlation Matrix Heatmap")

# Show the plot
plt.show()
```
