#### Q1. What is alpha? 
In statistics, alpha (α) represents the significance level of a hypothesis test. It is the threshold probability used to decide whether to reject the null hypothesis (H₀).

Alpha is the probability of making a Type I error — rejecting the null hypothesis when it is actually true.

#### Q7. Why do we use degrees of freedom?

Degrees of freedom (DOF) are used in statistics to indicate the number of independent values or quantities that can vary in an analysis without violating any constraints. They're crucial because they help adjust for bias when estimating population parameters from sample data.

We use DOF for:
- t-tests: to calculate the t-statistic and p-value.
- ANOVA: to determine the F-statistic and p-value.
- regression: to calculate the F-statistic and p-value.

#### Q10. What is a t-table?
A t-table (or Student's t-distribution table) is a reference table that shows the critical values of the t-distribution for different degrees of freedom (df) and confidence levels (or significance levels like 0.05, 0.01, etc.).
The t-table gives you the critical values to compare your calculated t-statistic against.

#### Q15. t-value vs t-test?

A t-test is a statistical test used to compare means and determine whether differences between groups are statistically significant. There are several types of t-tests:

- One-sample t-test: Compares the sample mean to a known value or population mean.
- Two-sample (independent) t-test: Compares means of two independent groups.
- Paired t-test: Compares means from the same group at different times (e.g., before and after).

The t-value is the test statistic you calculate during a t-test. It measures how far the sample mean is from the population mean (or the difference between two sample means), in units of standard error.

It’s calculated as:

𝑡 = Sample statistic − Hypothesized value / Standard error

A larger absolute t-value generally means a more significant difference.


#### Q16. t-test vs z-test?
t-tests and z-tests are both used to test hypotheses about population means, but they differ based on sample size and what you know about the population.

Use a z-test when:

- The sample size is large (typically 
𝑛 > 30)

- The population standard deviation (σ) is known.

- The population is normally distributed (or the sample size is large enough for the Central Limit Theorem to apply).


Use a t-test when:

The sample size is small (typically n<30).

The population standard deviation is unknown — you use the sample standard deviation (s) instead.

The population is normally distributed or approximately so.

#### Q2. What are the assumptions of the two-sample t-tests?

1. Independence of Observations
The two groups must be independent of each other.

Observations within each group must also be independent (no pairing or repeated measures).

Violation? → Use a paired t-test if samples are related.

2. Normality
The data in each group should be approximately normally distributed.

This is especially important when sample sizes are small (n < 30).

Large samples (by the Central Limit Theorem) are more forgiving of this.

3. Equal Variances (Homoscedasticity) – for standard (pooled) t-test
The two groups should have equal population variances.

Can be tested using:

Levene’s Test

F-test

Violation? → Use the Welch’s t-test, which does not assume equal variances.

## Q7. When do we use hypothesis testing? Why?
Hypothesis testing is used when we want to make data-driven decisions about a population based on a sample. It's a statistical method to determine whether there's enough evidence in a sample to support a specific claim about a population. We use hypothesis testing when we want to:

    1. ✅ Comparing Groups
Are two or more groups different?

Example: Is the average height of male and female students significantly different?

Test: t-test or ANOVA
    2. ✅ Evaluating Claims or Theories
Is a claim about a population true?

Example: A factory claims its light bulbs last 1000 hours. Do sample bulbs support that claim?

Test: One-sample t-test
    3. ✅ A/B Testing (Business / Marketing)
Does a new version perform better than the old?

Example: Is version B of a web page leading to more conversions than version A?

Test: Two-sample z-test or t-test for proportions
    4. ✅ Testing Correlation or Association
Is there a relationship between two variables?

Example: Are smoking and lung disease correlated?

Test: Chi-square test for categorical data, Correlation test for numerical data
    5. ✅ Medical Trials
Does a new treatment work better than the existing one?

Example: Does a new drug lower blood pressure more than the standard one?

Test: t-test, ANOVA, or non-parametric tests
